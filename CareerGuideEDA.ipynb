```python
import pandas as pd
```

```python
students = pd.read_csv("./Student data/student-scores.csv", encoding="ISO-8859-1")  
postings = pd.read_csv("./JobData/postings.csv", encoding="ISO-8859-1")  
job_skills = pd.read_csv("./JobData/job_skills.csv", encoding="ISO-8859-1")  
skills = pd.read_csv("./JobData/skills.csv", encoding="ISO-8859-1")  
salaries = pd.read_csv("./JobData/salaries.csv", encoding="ISO-8859-1")  
industries = pd.read_csv("./JobData/industries.csv", encoding="ISO-8859-1")  
job_industries = pd.read_csv("./JobData/job_industries.csv", encoding="ISO-8859-1")  
```

```python
students = pd.read_csv("./Student data/student-scores.csv", encoding="Windows-1252")  
```

## 1. Merge Job-Related Data

```python
job_skills_combined = pd.concat([job_skills, skills]).drop_duplicates()
```

```python
# Merge jobs with industries
jobs_with_industries = job_industries.merge(industries, on="industry_id", how="left")

# Merge jobs with salaries
jobs_with_salaries = salaries.merge(jobs_with_industries, on="job_id", how="left")
```

## 2. Relate Student Skills to Job Skills

```python

# Merge students with job skills
students_jobs = students.merge(job_skills_combined, left_on="career_aspiration", right_on="skill_abr", how="left")
```

```python
# Merge students with job and salary data
Finaldf = students_jobs.merge(jobs_with_salaries, on="job_id", how="left")
```

```python
career_skill_mapping = {
    'Lawyer': 'LGL',
    'Doctor': 'HCPR',
    'Government Officer': 'ADM',
    'Artist': 'ART',
    'Software Engineer': 'IT',
    'Teacher': 'EDU',
    'Business Owner': 'MGMT',
    'Scientist': 'SCI',
    'Banker': 'FIN',
    'Writer': 'WRT',
    'Accountant': 'ACCT',
    'Designer': 'DSGN',
    'Construction Engineer': 'ENG',
    'Game Developer': 'IT',
    'Stock Investor': 'FIN',
    'Real Estate Developer': 'SALE'
}

students['matched_skill'] = students['career_aspiration'].map(career_skill_mapping)
```

```python
salaries['job_id'] = salaries['job_id'].astype(str)
job_industries['job_id'] = job_industries['job_id'].astype(str)
job_skills['job_id'] = job_skills['job_id'].astype(str)
```

```python
# Merge students with job skills
students_jobs = students.merge(job_skills, left_on="matched_skill", right_on="skill_abr", how="left")
# Merge with job industries
students_jobs = students_jobs.merge(job_industries, on="job_id", how="left")
# Merge with salary data
final_data = students_jobs.merge(salaries, on="job_id", how="left")
```
