```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Initialize WebDriver
driver = webdriver.Chrome()

# Lists to store extracted data
role = []
company = []
region = []
pay_values = []
job_types = []
job_links = []
descriptions = []

# Iterate through multiple pages
for i in range(0, 50, 10):  
    link = f"https://in.indeed.com/jobs?q=&l=India&radius=25&start={i}"
    driver.get(link)
    time.sleep(5)  # Wait for page to load

    try:
        # Wait for the job listings to load
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "css-1faftfv")))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        job_listings = soup.find("ul", class_="css-1faftfv")  # Find job listings section

        if job_listings:
            for job in job_listings.find_all("li", class_="css-1ac2h1w"):
                role_txt = job.find("span").text if job.find("span") else "No role text"
                company_name = job.find("span", class_="css-1h7lukg").text if job.find("span", class_="css-1h7lukg") else "No company name"

                location_div = job.find("div", class_="css-1restlb")
                location = location_div.text.strip() if location_div else "No location"

                href_link = job.find("a")["href"] if job.find("a") else None

                # Construct full job link
                if href_link:
                    job_url = "https://in.indeed.com" + href_link
                    job_links.append(job_url)
                    print(f"Job Link: {job_url}")

                    # Open job details page
                    driver.get(job_url)
                    time.sleep(5)  # Give time for job page to load

                    soup1 = BeautifulSoup(driver.page_source, 'html.parser')

                    # Initialize variables to avoid errors
                    pay_value = "No pay info"
                    job_type = "No job type"
                    description = "No job description available"

                    # Extract Pay Value
                    pay_span = soup1.find("div", {"class": "css-1butozf eu4oa1w0"})
                    if pay_span:
                        pay_value = pay_span.text.strip()
                        print(f"Pay: {pay_value}")

                    # Extract Job Type
                    job_type_div = soup1.find("div", {"data-testid": "jobsearch-OtherJobDetailsContainer"})
                    if job_type_div:
                        job_type_span = job_type_div.find("span", class_="css-1kuy7z7")
                        if job_type_span:
                            job_type = job_type_span.text.strip()
                            print(f"Job Type: {job_type}")

                    # Extract Job Description
                    desc_div = soup1.find("div", class_="jobsearch-jobDescriptionText")
                    if desc_div:
                        description = desc_div.text.strip()
                        print(f"Description: {description[:200]}...")  # Print first 200 characters

                    # Store extracted data
                    pay_values.append(pay_value)
                    job_types.append(job_type)
                    descriptions.append(description)

                role.append(role_txt)
                company.append(company_name)
                region.append(location)

                # Print extracted details
                print(f"Role: {role_txt}")
                print(f"Company: {company_name}")
                print(f"Location: {location}")
                print("\n" + "="*50 + "\n")

        else:
            print("Couldn't find job listings on this page.")

    except Exception as e:
        print(f"Error on page {link}: {e}")

# Close the browser after scraping
driver.quit()
```
